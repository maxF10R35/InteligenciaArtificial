# PERCEPTRÃ“N

```python
class Perceptron:
    def __init__(self, pesos, sesgo, tp):# tp = tasa de aprendizaje
        self.pesos = pesos
        self.sesgo = sesgo
        self.tp = tp

    def funcion_activacion(self, z):
        if z >= 0:
            return 1
        else:
            return 0

    def ajustar_pesos(self, y_real, y_pred, entradas):
        nuevos_pesos = []
        for i in range(len(self.pesos)):
            nuevo_peso = self.pesos[i] + self.tp*(y_real - y_pred)*(entradas[i])
            nuevos_pesos.append(nuevo_peso)
        self.pesos = nuevos_pesos

    def ajustar_sesgo(self, y_real, y_pred):
        self.sesgo = self.sesgo + self.tp * (y_real - y_pred)

    def entrenar(self, entradas, salidas, epocas):
        for i in range(epocas):
            print(f' Esta es la epoca: {i}')
            y_uso = 0
            salidas_pred = []
            for entrada in entradas:
                print(f'Estas son las entradas {entrada}, estas son las salidas {salidas[y_uso]}')
                suma = (self.pesos[0] * entrada[0]) + (self.pesos[1] * entrada[1]) + self.sesgo
                salida = self.funcion_activacion(suma)
                salidas_pred.append(salida)
                salida_real = salidas[y_uso]
                y_uso += 1
                # ajustamos los pesos
                self.ajustar_pesos(salida_real, salida, entrada)
                self.ajustar_sesgo(salida_real, salida)
            print(f'El valor de los pesos en esta epoca es: {self.pesos}')
            print(f'El valor del sesgo en esta epoca es: {self.sesgo}')
            print(f'Las salidas predichas son: {salidas_pred}')

```

##AND

```python

#PERCEPTRON AND
pesos = [1,1]
sesgo = 1
tp = 0.1
entradas = [[1,1], [1,0], [0,1], [0,0]]
salidas = [1,0,0,0]

perceptron1 = Perceptron(pesos, sesgo, tp)

perceptron1.entrenar(entradas, salidas, 29)
```

```python
#PERCEPTRON OR
pesos = [1,1]
sesgo = 1
tp = 0.1
entradas = [[1,1], [1,0], [0,1], [0,0]]
salidas = [1,1,1,0]

perceptron1 = Perceptron(pesos, sesgo, tp)

perceptron1.entrenar(entradas, salidas, 29)
```
