```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Caminata aleatoria

```python
import random
```

```python
resultados = [0,]
inicio = 0
for i in range(10):
  lanzamiento = random.choice(['aguila','sol'])
  if lanzamiento == 'aguila':
    inicio += 1
  else:
    inicio -= 1
  resultados.append(inicio)
  print(lanzamiento, inicio)

print(resultados)
```

```python
indices = range(len(resultados))

plt.plot(resultados)
plt.scatter(indices, resultados, color="red")
plt.title("Caminata aleatoria")
plt.xlabel("Pasos")
plt.ylabel("Valor")
plt.grid(True)
plt.xticks(indices)

plt.show()
```
